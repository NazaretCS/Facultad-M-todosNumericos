### FORMULACIÓN DEL PROBLEMA
Se lleva a cavo mediante la abstracción de los datos del mundo real modelizando el problema en cuestión.
Modelizar un problema conyeva abstraer los datos del mundo real, simplificar la expreción, y visializar el contexto del problema.

Formular un _Modelo Matemático_ es definir las ecuaciones matematicas del modelo y los parametros que intervienen...

Un _modelo matemático_ es una ecuacion que muestra las caracteristicas de un sistema fisico:
**y = f(x, parámetros, fn.fuerza )**

* y = Variable dependiente: muestra el comportamineto o estado del sistema.
* x = Variables independientes: Determinan el comportamineto del sistema (Tiempo, espacio...)
* Parámetros: Costantes que muestran composición o propiedades del sistema.
* fn.fuerza: Influencias externa sobre el Sistema.



# CLASIFICACIÓN DE LOS ERRORES
__1. Errores del Modelo o del Problema__
__2. Errores del Calculo y Programación__
__3. Errores en los Datos__
__4. Errores de Truncamiento__
__5. Errores de Redondeo__



## 1. Errores del Modelo o del Problema
Son los más __Graves__, si modelamos mal, por ej. hacemos hipotesis incorrectas o demaciadas simplificaciones para llegar al modelo matemático, entonces no resolvemos el problema real planteado.


## 2. Errores del Calculo y Programación 
Son causados por el factor Humano, para evitarlos hay que revisar bien los programas que creamos.


## 3. Errores en los Datos
Los datos mayormente tienen errores asociados al instrumento con el cual se efectua la medición. **El Error en el resultado no puede ser menor que el error en los Datos**
Relacionado a estos errores esta el concepto de condicionamiento:

                                   __ ---> Error Pequeño en los resultados ---> Problema Bien Condicionado
    Error Pequeño en los Datos: [
                                   ------> Error Grande en los resultados ----> Problema Mal Condicionado

**Errores Sistematicos**: Son los que siempre tienen aproximadamente el mismo tamaño y magnitud, es decir tienen una causa constante.
**Errores Accidentales**: Son los relacionados a factores Aleatorios en la medición.
**Precisión Y Exactitud**: Una medida es Exacta cuando su valor es muy cercano al valor de verdad.
                           Una medida es Presisa cuando el instrumento que empleamos nos proporciona muchas sifras decimales, pero la medida puede no ser exacta.

## 4. Error de Truncamiento
Se origina al sustituir procesos infinitos por procesos finitos
Los errores de truncamiento causan mas inexactitud en los resultados.

## 5 Error de Redondeo
Se debe a que una maquina sólo puede representar cantidades con un número finito de dígitos. Se pueden distinguir dos tipos:
a). Errores de represetacion 
b). Error devido a los cálculos.





## FORMULAS PARA EL CALCULO DE ERRORES

**Sean x y x' un número y su valor aproximado**

e = x - x'                   Error Absoluto
er = ( x - x') / x           Error Relativo
er %  =  er * 100            Error Relativo Porcentaul
| x - x' | <= eps            Cota de Error  ------>  x = x' +- eps



**Sean x y x' e R a la n**

e = || x - x' ||            Error Absoluto
e = ( ||x-x'|| ) / ||x||    Error Relativo, si x <> 0




## REPRESENTACIÓN INTERNA  
### Punto Fijo:
Los Números se representan con un número fijo de cifras decimales.

### Punto Flotante:
Los Números se repsentan con un número fijo de dígitos significativos.


**Dígito Significativo**: Dado un numero x, es cualquier digito exepto los 0 a la izquierda del primer dígito diferente de cero y que solo sirven para fijar la posición del punto decimal: Ej. 1360, 1.360, 0.001360 tienen 4 dígitos significativos.



### REPRESENTACIÓN EN PUNTO FLOTANTE

Sea un Número x, representado en punto flotante en una base b
x = ( sign x)(.a1a2a3...at)b * b elevado e
ai = Díguitos en el sistema de la base, a1 <> 0 o ai = 0.
t = número de dígitos de la mantisa (determina la precisión)
.a1a2a3...at: mantiza normalizada por ser a1 > 0
e: exponente (determina el rango)

_A mayor mayor mantiza mayor precisión, y a mayor exponente mayor alncanze_


**Representación en Simple Precisión: 32bits**
    - 1 bit para el signo del número
    - 8 bits para el exponente
    - 23 bits para la mantiza


**Representación en Doble Precisión: 64bits**
    - 1 bit para el signo del número
    - 11 bits para el exponente
    - 52 bits para la mantisa.



**Overflow**: Resultado del cálculo mayor que el numero mas grande que se puede representar.
**Underflow**: Resultado del cálculo menor que el numero mas pequeño (no nulo) que se puede representar. Se concidera 0 al valor.


**ESTABILIDAD**
Si pequeños cambios en los datos producen pequeños cambios en los resultados diremos que es estable, caso contrario es inestable.
En algunos casos la estabilidad depende del conjunto de datos entonces decimos que es condicionalemnte estable:
_Inestabilidad Inherente_ es aquella propia del problema o sistema
_Inestabilidad inducida_ es la que se produce por usar un metodo equivocado para resolver un determinado problema.
