```python
from distutils.log import error
import numpy as np
import math

#------------DECLARACION DE LA FUNCION Y SU DERIVADA------------

def f(x):
    return 1/(x+1)

def f_1(x):
    return pow(x, 3) - x

def df(x):
    return 3*pow(x,2)-1

#------------METODO REGULAFALSI------------
def RegulaFalsi(a,b,eps,itMax):
    i = 0
    x = b - ((f(b)*(b-a))/(f(b)-f(a)))

    while i<=itMax and np.abs(f(x)>eps):
        
        if f(a)*f(x)>0:
            a=x
        else:
            b=x
        print('Valor en iteracion {}: {}'.format(i+1,x))
        x = b - ((f(b)*(b-a))/(f(b)-f(a)))
        i += 1
    
    if i > itMax:
        print('El metodo no converge en el numero de iteraciones dado. Intente nuevamente con un numero mayor de iteraciones')
    else:
        print('Raiz: {}'.format(x))

#------------METODO DE BISECCION------------
def Biseccion(a,b,eps,itMax):
    i = 0
    valMed = (a+b)/2
    
    while i<=itMax and np.abs(f(valMed))>eps:
        error = np.abs(a-valMed)
        if f(a)*f(valMed)>0:
            a = valMed
        else:
            b = valMed
        print('Valor en iteracion {}: {} / error: {}'.format(i+1,valMed, error))
        valMed = (a+b)/2
        i += 1
    
    if i > itMax:
        print('El metodo no converge en el numero de iteraciones dado. Intente nuevamente con un numero mayor de iteraciones')
    else:
        print('Raiz: {}'.format(valMed))

#------------METODO DE LA SECANTE------------
def Secante(a,b,eps,itMax):
    i = 0 
    x_n_1 = a
    x_n = b
    #x_n1 = 2*x_n
    x_n1 = x_n - (f(x_n) * ((x_n - x_n_1)/(f(x_n) - f(x_n_1))))

    while i<=itMax and np.abs(x_n1-x_n)>eps:
        x_n_1 = x_n
        x_n = x_n1
        error = np.abs(x_n_1 - x_n)
        print('Valor en iteracion {}: {} / error: {}'.format(i+1,x_n1, error))
        x_n1 = x_n - (f(x_n) * ((x_n - x_n_1)/(f(x_n) - f(x_n_1))))
        i+=1
    
    if i>itMax:
        print('El metodo no converge en el numero de iteraciones dado. Intente nuevamente con un numero mayor de iteraciones')
    else:
        print('Raiz: {}'.format(x_n1))
    
#------------METODO NEWTON-RAPHSON------------
def NewtonRaphson(x,eps,itMax):
    i=0
    #x_1 = x + 2*eps
    x_1 = x - (f_1(x)/df(x))

    while i<=itMax and np.abs(x_1-x)>eps:
        error = np.abs(x_1-x)
        print('Valor en iteracion {}: {} / error: {}'.format(i+1,x_1, error))
        x = x_1
        x_1 = x - (f_1(x)/df(x))
        i+=1
    
    if i>itMax:
        print('El metodo no converge en el numero de iteraciones dado. Intente nuevamente con un numero mayor de iteraciones')
    else:
        print('Raiz: {}'.format(x_1))

#------------Metodo de Iteracion de Punto Fijo------------
def PuntoFijo(x, eps, itMax):
    for i in range(itMax):
        x_1 = f(x)
        if np.abs(x_1-x)<eps:
            error = np.abs(x_1-x)
            print('Iteracion donde converge {}, Valor del punto fijo: {} // ERROR: {}'.format(i+1, x_1, error))
            return
        error = np.abs(x_1-x)
        print('Valor de iteracion {}: {} // ERROR: {}'.format(i+1, x_1, error))
        x = x_1
        
    if i+1==itMax:
            print('El metodo no converge en el numero de iteraciones dado. Intente nuevamente con un numero mayor de iteraciones')




#------------METODO REGULAFALSI------------
print('\n\n')

print('====Metodo Regula Falsi====')
RegulaFalsi(0.5, 1.5, 0.00001, 150)

#------------METODO DE BISECCION------------
print('\n\n')

print('====Metodo de Biseccion====')
Biseccion(0.5, 1.5, 0.00001, 150)

#------------METODO DE LA SECANTE------------
print('\n\n')

print('====Metodo de la Secante====')
Secante(0.5, 1.5, 0.00001, 150)

#------------METODO NEWTON-RAPHSON------------
print('\n\n')

print('====Metodo de Newton-Raphson====')
NewtonRaphson(np.sqrt(1/5), 0.0001, 150)

#------------Metodo de Iteracion de Punto Fijo------------
print('\n\n')

print('===Metodo de Iteracion de Punto Fijo===')
PuntoFijo(1, 0.001, 150)
```

```python
"""
Otra copiada
"""

#Metodo de la secante
import numpy as np

def f(x):
    return (x**(-1)) - np.tan(x)

def secante(x0, x1, tol, max_iter):
    # x0 y x1 son los dos puntos iniciales
    # tol es la tolerancia para el criterio de parada
    # max_iter es el nmero mximo de iteraciones permitidas

    # Inicializa las variables
    iteracion = 0
    error = tol + 1
    fx0 = f(x0)
    fx1 = f(x1)

    # Itera hasta alcanzar la tolerancia o el nmero mximo de iteraciones
    while error > tol and iteracion < max_iter:
        x2 = x1 - fx1*(x1 - x0)/(fx1 - fx0) # Calcula el siguiente punto de la secante
        fx2 = f(x2) # Evala la funcin en el siguiente punto
        error = abs(x2 - x1) # Calcula el error absoluto
        x0 = x1 # Actualiza los valores de x0 y x1
        fx0 = fx1
        x1 = x2
        fx1 = fx2
        iteracion += 1 # Actualiza el contador de iteraciones

    # Devuelve el valor aproximado de la raz y el nmero de iteraciones realizadas
    return x2, iteracion

# Ejemplo de uso
raiz_aprox, iteraciones = secante(0.5, 1.5, 0.00001, 100)
print("La raz aproximada es: {:.6f}".format(raiz_aprox))
print("Nmero de iteraciones realizadas: {}".format(iteraciones))



#metodo de regula falsi
```

```python
# Metodo de NewtonRaphson


import numpy as np
from math import sqrt

def f(x):
    return (x**(3)) - x

def d(x):
    return (3*x**2 - 1)


#------------   METODO DE LA NEWTON-RAPHSON   ------------

def NewtonRaphson(xn, eps, MaxIt):
    iteracion = 0 
    xn_1 = (xn - f(xn)/d(xn))
    #xn = b
    
    while (iteracion <= MaxIt and np.abs(xn_1 - xn) > eps):
        iteracion += 1
        
        #xn = xn1
        error = np.abs(xn_1 - xn)
        print('Valor en iteracion ', iteracion, ": ", xn_1, " Con un error de: ", error)
        xn = xn_1 
        xn_1 = xn - (f(xn)/d(xn))
        
    
    if iteracion > MaxIt:
        print('El mtodo no converge en ', MaxIt, ' Iteraciones')
    else:
        print('El mtodo a convergido\nRaiz: ',xn_1)

        
    

#------------   METODO DE NEWTON-RAPHSON   ------------
print('\n\n__________   Mtodo De NEWTON-RAPHSON   __________\n')
#NewtonRaphson(sqrt((1/5)), 1.0E-3, 150)
NewtonRaphson((0.1), 1.0E-3, 150)

```
