```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 1.01*np.exp(4*x) - 4.62*np.exp(3*x) - 3.11*np.exp(2*x) + 12.2*np.exp(x) - 1.99

x = np.linspace(-1, 2, 5000)
y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gr√°fica de f(x)')
plt.grid()
plt.show()

```
