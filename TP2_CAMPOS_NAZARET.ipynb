```python
# Punto 1 D - Compare los resultados y saque concluciones.

"""

    Ambos metodos , en este caso, convergen. La diferencia enta en como lo hacen y en el tiempo o en el n° de iteraciones 
    en que lo hacen.
    En el metodo de biseccion la convergencia es lenta y el error siempre se va reduciendo a la mitad, por lo que podriamos hablar
    de una convergencia lineal. La convergencia de este metodo tambien se ve afectada por la naturaleza de la funcion a tratar
    por ejemplo podriamos decir que tenemos asegurada una convergencia lenta si la funcion es muy suave (es decir si no tiene una
    curvatura muy pronunciada en la region de la raiz) o si la region que contiene a la raiz es muy grande.
    La convergencia del metodo de la secante, en este caso y en general, es mas rapida que la del metodo de biseccion (Tiene una 
    orden de convergencia de 1.6). Sin envargo hay ocaciones en las que la velocidad de convergencia o incluso la convergencia en 
    si se ven comprometidas, por ejemplo si la funcion presenta una raiz multiple o si es casi constante cerca de la raiz 
    
    
    Dato aparte: 
        Cuando se habla de un orden de convergencia se esta haciando referencia a la rapidez con la que un metodo converge a la 
        solucion exacta. El orden de convergencia se expresa como una potencia del error de aproximacion en cada iteracion. Un orden
        de error del 1.6 significa que en cada iteracion el error de aproxiacion disminuye aproximadamente a la 1.6 potencia, osea
        que si en una iteracion el error es de 0.1, en la siguiente sera de aproximadamente (1)**1.6
        

"""
```

```python
# Punto 1 e: Programe el método de Regula Falsi y el método de la Secante.

#------------   METODO REGULAFALSI   ------------
# while (iteracion < MaxIt) and np.abs(f(x)>eps):
# while (iteracion < MaxIt) and (np.abs(a-b) > eps):

import numpy as np

def f(x):
    return (x**(-1)) - np.tan(x)
    
def RegulaFalsi(a, b, eps, MaxIt):
    if( f(a) * f(b) > 0):
        print("\nLa Raiz no se encuentra en el intervalo")
        iteracion = 0
        x = "Raiz no encontrada"
    else:
        iteracion = 0
        x = b - f(b)*(b-a)/(f(b)-f(a))
        while (iteracion < MaxIt) and (np.abs(a-b) > eps):
        #while (iteracion < MaxIt) and np.abs(f(x)>eps):
            iteracion += 1
            if (f(a)*f(x)) > 0:
                a = x
            else:
                b = x
            error = abs(a-b)
            
            x = b - (f(b)*(b-a))/(f(b)-f(a))
            print('Valor en iteracion {}: {} con un error de: {}, a: {}, b: {}'.format(iteracion,x,error,a, b))
           
    if (iteracion >= MaxIt):
        print("No converge en ", MaxIt, " Iteraciones")
    else:
        print("Convergio al valor de: ",x)
        


        
        
#------------   METODO DE LA SECANTE   ------------
def Secante(a, b, eps, MaxIt):
    iteracion = 0 
    xn_1 = a
    xn = b
    xn1 = xn - f(xn) * (xn - xn_1) / (f(xn) - f(xn_1))

    #while (iteracion <= MaxIt and np.abs(xn1 - xn) > eps):
    while (iteracion <= MaxIt and np.abs(xn1 - xn) > eps):
        iteracion += 1
        xn_1 = xn
        xn = xn1
        error = np.abs(xn_1 - xn)
        print('Valor en iteracion ', iteracion, ": ", xn1, " Con un error de: ", error)
        xn1 = xn - f(xn) * (xn - xn_1) / (f(xn) - f(xn_1))
        
    
    if iteracion > MaxIt:
        print('El método no converge en ', MaxIt)
    else:
        print('El método a convergido\nRaiz: ', xn1)

        
        
        
        
#___________    METODO DE BISECCION    _______________  (No lo pedia, pero lo programe por las dudas)

def biseccion(a, b, eps, maxiteraciones):
    if (f(a)*f(b) > 0):
        print("La raiz no se encuentra en el interbalo")
    else:
        iteracion = 0
        p = (a + b) / 2
        while (iteracion < maxiteraciones and abs(a-b)>eps):
            print("Iteracion N°: ",iteracion, " valor de p: ",p," con un error de: ",abs(a-b))
            iteracion += 1
            if (f(a) * f(p) > 0):
                a = p
            else:
                b = p
            p = (a +b) / 2
    if (iteracion >= maxiteraciones):
        print("El metodo no converge en: ", maxiteraciones)
    else:
        print("El metodo a convergido a la raiz: ", p)
        

        
        
        
#------------   METODO REGULAFALSI   ------------
print('\n\n__________   Método Regula Falsi   __________\n')
RegulaFalsi(0.5, 1.5, 1.0E-3, 1500)


#------------   METODO DE LA SECANTE   ------------
print('\n\n \n\n__________   Método de la Secante   __________\n')
Secante(0.5, 1.5, 1.0E-3, 150)


#------------   METODO DE BISECCION   ------------
print('\n\n \n\n__________   Método de Biseccion   __________\n')
biseccion(0.5, 1.5, 1.0E-3, 150)




"""
    ¿Cuál es la diferencia entre ambos métodos?
    
    La principal diferencia es que el metodo de la secante si converge a diferencia del Regula-Falsi (en el intervalo dado,
    con una tolerancia de 0.1E-3, y un maximo de iteraciones de 150). Esto sucede por la forma de la funcion dada y por la forma
    de trabajo del metodo en cuestion.
    El metodo de Regula-Falsi tambien puede sufrir de estancamiento cuando hay raices multiples o cuando la funcion es casi
    horisontal en el interbalo dado (como es el caso)
    
    
"""
```

```python
# Punto 2: Implementacion de Newton–Raphso

import numpy as np
from math import sqrt

def f(x):
    return (x**(3)) - x

def d(x):
    return (3*x**2 - 1)


#------------   METODO DE LA NEWTON-RAPHSON   ------------

def NewtonRaphson(xn, eps, MaxIt):
    iteracion = 0 
    xn_1 = (xn - f(xn)/d(xn))
    #xn = b
    
    while (iteracion <= MaxIt and np.abs(xn_1 - xn) > eps):
        iteracion += 1
        
        #xn = xn1
        error = np.abs(xn_1 - xn)
        print('Valor en iteracion ', iteracion, ": ", xn_1, " Con un error de: ", error)
        xn = xn_1 
        xn_1 = xn - (f(xn)/d(xn))
        
    
    if iteracion > MaxIt:
        print('El método no converge en ', MaxIt, ' Iteraciones')
    else:
        print('El método a convergido\nRaiz: ',xn_1)

        
    

#------------   METODO DE NEWTON-RAPHSON   ------------
print('\n\n__________   Método De NEWTON-RAPHSON   __________\n')
NewtonRaphson(sqrt((1/5)), 1.0E-3, 150)
#NewtonRaphson((0.1), 1.0E-3, 150)


"""


    ¿Qué ocurre si ejecuta el algoritmo con la función f(x) = x3 − x , con el valor inicial x0= √1/5?
        Los resultados obtenidos demuestran un comportamiento ciclico entre los valores -0.4472135954999579 y 0.4472135954999579
        osea que el metodo esta diverguiendo en lugar de converger esto es causado por el valor inicial que elejimos (X0 = sqrt((1/5))
        esta muuy cerca de la raiz x0, y como resultado obtenemos este conportamiento oscilante sin converger.
        Si modificamos el valor de X0 el metodo si converge, en algunos casos con muchas iteraciones, en otros con unas pocas.
        Una eleccion inicial adecuada es crucial para el exito del metodo de Newton-Raphson.
        
        
    
"""
```

```python
# Punto 3

"""
    Dificultades que se podrian presentar:
    --------------------------------------
        
        * No llegar a converger por una mala eleccion del valor inicial.
        * Que la funcion tenga raices multiples.
        * Que la derivada de la funcion sea. (imposivilita el calculo de f(x) / d(x)
        * Que el valor inicial no se encuentre en el dominio de la fn
        * Al usar la derivada de la funcion requiere de su calculo, lo cual representa una desventaja frente a otros metodos
          al momento de hablar de costo computacional.
          
"""


"""
    Compare y Saque conclusiones
    ----------------------------
        
        Frente  aun mismo valor inicial (X0 = 2) no todas las funciones converguen (1), y las que si lo hacen varian mucho en 
        la cantidad de iteraciones que utilizan para buscar el valor de la raiz (2)
        
        * (1): Por ej. la fn f4, con X0 = 2 conforme pasan las iteraciones se llega a un punto donde la derivada se hace cero, 
          por lo que se nos imposivilita calcular en Xn siguiente, y por lo tanto el metodo diverge. Esto es algo particular
          del metodo, ya que al utilizar la derivada como denominador sabemos que tenemos que andar "esquivando" estos valores.
          
        * (2): El hecho de que unas funciones convergan mas rapido que otras tambien esta relacionado con el valor inicial
          ya que el metodo de newton-Raphson tiene la particularidad de que si el valor inicial esta proximo a la raiz convergira 
          mas rapido y a la raiz mas proxima.

"""


import numpy as np
import math

# Funcion a y su derivada: f(x) = sen(x) - ½ √ x
def f1(x):
    return (math.sin(x) - 0.5 * math.sqrt(x))

def d1(x):
    return (math.cos(x) - (1 / (2 * math.sqrt(x))))



# Funcion b y su derivada:  tg(x) -0.5x
def f2(x):
    return ( math.tan(x) - 0.5 * x )

def d2(x):
    return (math.pow(math.cos(x), -2) - 0.5)



# Funcion c y su derivada: x 10 -1
def f3(x):
    return ( x**10 -1 )

def d3(x):
    return( 10 * (x ** 9) )



# Funcion d y su derivada: x**3 - 5 x**2 + 7 x - 3
def f4(x):
    return( x**3 - 5*x**2 + 7*x - 3 )

def d4(x):
    return ( 3 * (x ** 2) - 10 * x + 7 )



# Funcion e y su derivada: x**2 – x – 2,5
def f5(x):
    return ( x**2 - x - 2.5 )

def d5(x):
    return ( 2 * x - 1 )

#------------   METODO DE LA NEWTON-RAPHSON   ------------

def NewtonRaphson(xn, eps, MaxIt, f, d):
    iteracion = 0 
    xn_1 = (xn - f(xn)/d(xn))
    #xn = b
    
    while (iteracion <= MaxIt and np.abs(xn_1 - xn) > eps):
        iteracion += 1
        
        #xn = xn1
        error = np.abs(xn_1 - xn)
        print('Valor en iteracion ', iteracion, ": ", xn_1, " Con un error de: ", error)
        xn = xn_1 
        #print("fxn=",f(xn),"dx=",d(xn))
        xn_1 = xn - (f(xn)/d(xn))
        
    
    if iteracion > MaxIt:
        print('El método no converge en ', MaxIt, ' Iteraciones')
    else:
        print('El método a convergido\nRaiz: ',xn_1)

        


#------------   METODO DE NEWTON-RAPHSON   ------------
print('\n\n__________   Método De NEWTON-RAPHSON CON F1   __________\n')
#NewtonRaphson(sqrt((1/5)), 1.0E-3, 150)
NewtonRaphson((2), 1.0E-3, 150, f1, d1) 


#------------   METODO DE NEWTON-RAPHSON   ------------
print('\n\n__________   Método De NEWTON-RAPHSON CON F2   __________\n')
#NewtonRaphson(sqrt((1/5)), 1.0E-3, 150)
NewtonRaphson((2), 1.0E-3, 150, f2, d2) 


#------------   METODO DE NEWTON-RAPHSON   ------------
print('\n\n__________   Método De NEWTON-RAPHSON CON F3   __________\n')
#NewtonRaphson(sqrt((1/5)), 1.0E-3, 150)
NewtonRaphson((2), 1.0E-3, 150, f3, d3) 



#------------   METODO DE NEWTON-RAPHSON   ------------
print('\n\n__________   Método De NEWTON-RAPHSON CON F4 (1)  __________\n')
#NewtonRaphson(sqrt((1/5)), 1.0E-3, 150)
NewtonRaphson((2.5), 1.0E-3, 150, f4, d4) 

#------------   METODO DE NEWTON-RAPHSON   ------------
print('\n\n__________   Método De NEWTON-RAPHSON CON F4 (2)   __________\n')
#NewtonRaphson(sqrt((1/5)), 1.0E-3, 150)
NewtonRaphson((0.5), 1.0E-3, 150, f4, d4) 



#------------   METODO DE NEWTON-RAPHSON   ------------
print('\n\n__________   Método De NEWTON-RAPHSON CON F5   __________\n')
#NewtonRaphson(sqrt((1/5)), 1.0E-3, 150)
NewtonRaphson((2), 1.0E-3, 150, f5, d5) 




```
